Question 2
Using the fielding table, group players into three groups based on their position: label players with position OF as "Outfield", those with position "SS", "1B", "2B", and "3B" as "Infield", and those with position "P" or "C" as "Battery". Determine the number of putouts made by each of these three groups in 2016.

```python
posgroup = {'Battery':['P','C'],'Infield':['SS','1B','2B','3B'],'Outfield':['OF']}
def grouppos(pos):
    if pos in posgroup['Battery']:
        return 'Battery'
    elif pos in posgroup['Infield']:
        return 'Infield'
    elif pos in posgroup['Outfield']:
        return 'Outfield'

```

```python
lahman['fielding']['PosGroup'] = lahman['fielding']['pos'].apply(grouppos)
```

```python
lahman['fielding'][lahman['fielding']['yearid']==2016].loc[:,['PosGroup','po']].groupby('PosGroup').agg('sum')
```
